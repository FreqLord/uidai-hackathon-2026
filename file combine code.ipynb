```python
import pandas as pd
import glob
```

```python
# Method 1: Using pandas.concat()
def combine_csv_files(files, output_file='api_data_aadhar_enrolment.csv'):
    """Combine CSV files with same structure"""
    df_list = []
    
    for file in files:
        df = pd.read_csv(file)
        df_list.append(df)
    
    combined_df = pd.concat(df_list, ignore_index=True)
    combined_df.to_csv(output_file, index=False)
    print(f"Combined {len(files)} files into {output_file}")
    return combined_df

# Usage
files = ['api_data_aadhar_enrolment_0_500000.csv', 'api_data_aadhar_enrolment_500000_1000000.csv', 'api_data_aadhar_enrolment_1000000_1006029.csv']
combine_csv_files(files)
```
